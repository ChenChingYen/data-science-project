# Calculate User Preferences Data based on Past Behaviors

the **"rating"** score of each category is calculated based on the frequency of the user interacted with certain categories of product.

```python
# from collections import defaultdict

# user_preferences = defaultdict(int)
# for index, row in result_df.iterrows():
#   for index, category in enumerate(row["category_code"].split(".")):
#     user_preferences[category] += (row["rating"] * (index+1))
```

```python
from collections import defaultdict
```

```python
user_preferences = defaultdict(int)
for index, row in result_df.iterrows():
  for index, category in enumerate(row["category_code"].split(".")):
    user_preferences[category] += (row["rating"] * (index+1))
```

```python
user_preferences
```

```python
sorted(user_preferences.items(), key=lambda x:x[1], reverse=True)
```

# Normalization

the "rating" score is normalise to ensure that the comparison between products is standard

```python
min_value = min(list(user_preferences.values()))
max_value = max(list(user_preferences.values()))
for tag, score in user_preferences.items():
  user_preferences[tag] = (score - min_value) / (max_value - min_value)
```

# Calculate Popular Products under each Category

this step is used to identify the popularity of various products of the same category. The products with higher popularity have a higher chance to be recommended towards more customers.

```python
tag_hots = defaultdict(dict)
```

```python
chunk_size = 1000000
for filename in filename_list:

  # load dataset in chunks
  chunks = pd.read_csv(file_path_prefix + filename, chunksize=chunk_size)
  print(f"reading {filename}")

  for index, chunk in enumerate(chunks):
    for index, row in chunk.iterrows():
      product_id = row["product_id"]
      brand = row["brand"]
      # print(f"{index} {product_id} {brand} : ", end="")
      for category in row["category_code"].split("."):
        # print(category, end=", ")
        if product_id not in tag_hots[category]:
          tag_hots[category][product_id] = 0.0
        tag_hots[category][product_id] += row["rating"]
        ###
      # print()

  print(f"{filename} completed")
```

```python
sorted(tag_hots["sofa"].items(), key=lambda x:x[1], reverse=True)[:10]
```

```python
for category in tag_hots:
  min_value = min(list(tag_hots[category].values()))
  max_value = max(list(tag_hots[category].values()))
  for product_id, rating in tag_hots[category].items():
    if max_value - min_value != 0:
      tag_hots[category][product_id] =
       (rating - min_value) / (max_value - min_value)
    else:
      tag_hots[category][product_id] = 1
```

```python
sorted(tag_hots["sofa"].items(), key=lambda x:x[1], reverse=True)[:10]
```

```python
# export calculation results in csv
import csv

output_file_path = file_path_prefix + 'popular_product_csv.csv'
with open(output_file_path, 'w', newline='') as csvfile:

  csv_writer = csv.writer(csvfile)
  csv_writer.writerow(['category', 'product_id', 'rating'])

  for category, products in tag_hots.items():
    for product, rating in products.items():
      csv_writer.writerow([category, product, rating])
```

```python
# export calculation results in JSON
import json

json_file_path = file_path_prefix + 'popular_product_json.json'

with open(json_file_path, 'w') as jsonfile:
  json.dump(tag_hots, jsonfile, indent=2)
```

# Top 10 Products based on User's Preferences

the likelihood of the selected user interested in that specific product is calculated through the multiplication between the interest of the customer in that category and the popularity of that product.

```python
target_products = defaultdict(float)
```

```python
for user_category, user_score in user_preferences.items():
  for product, product_score in tag_hots[user_category].items():
    target_products[product] += user_score * product_score
```
